```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shampoo_sales.csv',
                index_col='Month',
    parse_dates=['Month'],)

df.head(10)
```

```python
# Nastavení stylu grafu
sns.set(style="whitegrid")

# Vytvoření grafu
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Month', y='Sales', palette='viridis', linewidth=2.5)

# Zvýraznění bodů dat
sns.scatterplot(data=df, x='Month', y='Sales', color='blue', s=50)

# Úprava velikosti a stylu fontu
plt.xlabel('Datum', fontsize=14, fontweight='bold')
plt.ylabel('Hodnota', fontsize=14, fontweight='bold')
plt.title('Vylepšená časová řada', fontsize=16, fontweight='bold')

# Zobrazení grafu
plt.show()
```
